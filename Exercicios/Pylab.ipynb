```python
# O Pylab combina funcionalidades do pyplot com funcionalidades do numpy
from pylab import *
%matplotlib inline
```

```python
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

# Definindo os eixos 

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de linha');
```

```python
# Gráfico com duas figuras 
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

# Definindo os eixos 

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Figura principal')

axes2.plot(x, y, 'b')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Figura secundária');
```

```python
# Figura com gráfico paralelo 
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')
fig.tight_layout()
```
